## Make UCI dataset

```python
import pandas as pd
```

```python
names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
         'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
         'hours-per-week', 'native-country', 'income']


df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', names=names)
df['income'] = df['income'].apply(lambda x: 1 if x.strip() == '>50K' else 0)
```

```python
df.head()
```

```python
df_train = df.iloc[:170, :]
df_test = df.iloc[170:, :]

df_train.to_csv('ucidata_train.csv', index=False)
df_test.to_csv('ucidata_test.csv', index=False)
```
