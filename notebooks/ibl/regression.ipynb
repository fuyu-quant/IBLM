```python
import sys
sys.path.append("../..")

"""
from importlib import reload
import src.iblm.ibl as ibl_module
reload(ibl_module)
from src.iblm.ibl import IBLModel
"""

from iblm import IBLModel, IBLBaggingModel


import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error, r2_score
```

```python
num_train_data = 300 # 50〜300
num_test_data = 1000
sample = num_train_data + num_test_data
n_informative = 2
n_redundant = 1
n_features = n_informative + n_redundant
weights = [0.5, 0.5]
flip_y=0
seed = 3655

x, y = make_regression(
    n_samples=sample,
    n_features=n_features,
    n_informative=n_informative,
    noise=0.1,
    random_state=seed
    )

x = np.round(x, decimals=3)
x = pd.DataFrame(x)
x.columns = ['x1', 'x2', 'x3']
y = np.round(y, decimals=2)

x_test = x[0:num_test_data]
x_train = x[num_test_data:]

y_test = y[:num_test_data]
y_train = y[num_test_data:]

x_train.head()
```

```python
ibl = IBLModel(
    api_type="claude",
    #api_type="gemini",
    #api_key="openai",
    model_name="claude-3-opus-20240229",
    #model_name="gemini-pro",
    #model_name="gpt-4-0125-preview",
    objective="regression"
    #objective="binary"
    )
```

```python
model = ibl.fit(x_train, y_train)
print(model)
```

```python
y_pred = ibl.predict(x_test)
ibl.evaluate(y_test, y_pred)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R^2スコア:", r2)
```
