# Circle

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

import pandas as pd
```

```python
sample = 2000

X, y = make_circles(n_samples=sample, noise=0.05, random_state=3655)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
df = pd.DataFrame(data=X, columns=['Feature_1', 'Feature_2']).round(3)
df['Target'] = y
```

```python
num_train = 200
sample_num = int(num_train/2)

df_1 = df[df['Target'] == 1].sample(n=sample_num, random_state=3655)
df_0 = df[df['Target'] == 0].sample(n=sample_num, random_state=3655)
df_train = pd.concat([df_1, df_0])
df_train = df_train.sample(frac=1, random_state=3655).reset_index(drop=True)
df_test = df.drop(df_train.index)

path = '../../data/circle/'
df_train.to_csv(path + f'circle_{num_train}_train.csv', index=False)
df_test.to_csv(path + f'circle_{num_train}_test.csv', index=False)
```
