# Circle

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

import pandas as pd
```

```python
sample = 600

X, y = make_circles(n_samples=600, noise=0.05, random_state=3655)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
df = pd.DataFrame(data=X, columns=['Feature_1', 'Feature_2']).round(4)
df['Target'] = y
```

```python
sample_num = int(sample/4)

df_1 = df[df['Target'] == 1].sample(n=sample_num, random_state=3655)
df_0 = df[df['Target'] == 0].sample(n=sample_num, random_state=3655)
df_train = pd.concat([df_1, df_0])
df_test = df.drop(df_train.index)

path = '../../data/circle/'
df_train.to_csv(path + f'circle_{sample}_train.csv', index=False)
df_test.to_csv(path + f'circle_{sample}_test.csv', index=False)
```
