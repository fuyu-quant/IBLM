# Moon

```python
import pandas as pd
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
sample = 600

X, y = make_moons(n_samples = sample, noise=0.1, random_state = 3655)


plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.show()
```

```python
df = pd.DataFrame(data=X, columns=['Feature_1', 'Feature_2'])
df['Target'] = y
```

```python
sample_num = int(sample/4)

df_1 = df[df['Target'] == 1].sample(n=sample_num, random_state=3655)
df_0 = df[df['Target'] == 0].sample(n=sample_num, random_state=3655)
df_train = pd.concat([df_1, df_0])
df_test = df.drop(df_train.index)

path = '../../data/moon/'
df_train.to_csv(path + f'moon_{sample}_train.csv', index=False)
df_test.to_csv(path + f'moon_{sample}_test.csv', index=False)
```
