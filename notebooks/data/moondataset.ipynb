# Moon

```python
import pandas as pd
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
sample = 1000

X, y = make_moons(n_samples = sample, noise=0.05, random_state = 3655)


plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.show()
```

```python
df = pd.DataFrame(data=X, columns=['Feature_1', 'Feature_2']).round(3)
df['Target'] = y
```

```python
len(df)
```

```python
num_train = 400
sample_num = int(num_train/2)

df_1 = df[df['Target'] == 1].sample(n=sample_num, random_state=3655)
df_0 = df[df['Target'] == 0].sample(n=sample_num, random_state=3655)
df_train = pd.concat([df_1, df_0])
#df_train = df_train.sample(frac=1, random_state=3655).reset_index(drop=True)
df_test = df.drop(df_train.index)

path = '../../data/moon/'
df_train.to_csv(path + f'moon_{num_train}_train.csv', index=False)
df_test.to_csv(path + f'moon_{num_train}_test.csv', index=False)
```
